```python
import os
import subprocess
import glob
from pathlib import Path 
```

```python
og_source_dir = '/media/Data/GoogleDriveNtust/dissertation_01/work_2020_04/IMP'
work_dir = '/media/Data/GoogleDriveNtust/dissertation_01/workFinal/workDir'
subjectLimit = 50
```

```python
#Creates BIDS compatible sourcedata directory
try:
    source_path = fr'{work_dir}/test_dataset/sourcedata'
    os.makedirs(source_path)
except OSError:
    print ("Creation of the directory %s failed" % source_path)
else:
    print ("Successfully created the directory %s " % source_path)
```

```python
#Creates BIDS compatible rawdata directory
try:
    raw_path = fr'{work_dir}/test_dataset/rawdata'
    os.makedirs(raw_path)
except OSError:
    print ("Creation of the directory %s failed" % raw_path)
else:
    print ("Successfully created the directory %s " % raw_path)
```

```python
#Creates BIDS compatible derivatives directory
try:
    derivatives_path = fr'{work_dir}/test_dataset/derivatives'
    os.makedirs(derivatives_path)
except OSError:
    print ("Creation of the directory %s failed" % derivatives_path)
else:
    print ("Successfully created the directory %s " % derivatives_path)
```

```python
results_dir = Path(f'{work_dir}/registration_results')
try:
    os.mkdir(results_dir)
except OSError:
    print(f'the directory {results_dir} could not be created')
else:
    print(f'the directory {results_dir}')
```

```python
source_data_subs = sorted(os.listdir(og_source_dir))
source_data_subs
```

```python
for i in range(subjectLimit):
    print(source_data_subs[i])
    
    i_formated = str(i+1).zfill(2)
    print(i_formated)
```

```python
for i in range(subjectLimit):
    #Transfers Source Subjects to BIDS source directory
    in_dir_path = f'{og_source_dir}/{source_data_subs[i]}/MPRAGE'
    print(in_dir_path)
    
    try:
        source_dir_path = f'{source_path}/{source_data_subs[i]}'
        os.makedirs(source_dir_path)
    except OSError:
        print ("Creation of the directory %s failed" % source_dir_path)
    else:
        print ("Successfully created the directory %s " % source_dir_path) 
    
    
    rc = subprocess.run(['cp', '-aR',f'{in_dir_path}', f'{source_dir_path}'])
    
    print('\n')
    
    #Creates BIDS compatible subject directories at raw directory
    i_formated = str(i+1).zfill(2)
    sub_raw_path = fr'{raw_path}/sub-{i_formated}'
    try:
        os.mkdir(sub_raw_path)
    except OSError:
        print (f'Creation of the directory {sub_raw_path} failed')
    else:
        print (f'Successfully created the directory {sub_raw_path}')
        
    print('\n')
    
    #transfer dicom images to compressed nifit format using dcm2niix
    
    output_dir = sub_raw_path
    print(output_dir)
    
    print('\n')
    
    rc = subprocess.run(['dcm2niix', '-o', f'{output_dir}', f'{source_dir_path}'])    
```

```python
pl0_dir = f'{derivatives_path}/pipeline_0'

try:
    os.mkdir(pl0_dir)
except OSError:
    print(f'the directory {pl0_dir} could not be created')
else:
    print(f'the directory {pl0_dir} was be created succesfully')
```

```python
list_rawdata_dir = os.listdir(raw_path)
```

```python
for sub_num in list_rawdata_dir:
    
    #pl0s_dir is pipeline_0 subject directories
    pl0s_dir = f'{pl0_dir}/{sub_num}'
    try:
        os.mkdir(pl0s_dir)
    except OSError:
        print(f'the directory {pl0s_dir} could not be created')
    else:
        print(f'the directory {pl0s_dir} was created succesfully')
    
    os.chdir(pl0s_dir)
    
    #snr_path is subject number raw path
    sub_num_raw_path_list = glob.glob(f'{raw_path}/{sub_num}/*.nii')
    snr_path = sub_num_raw_path_list[0]
    print(snr_path)
    
    
    
    print('\n')
    
    rc = subprocess.run(['bet', f'{snr_path}', f'bet_{sub_num}'])
    
    rc = subprocess.run(['gzip', '-d', f'bet_{sub_num}.nii.gz'])
```
